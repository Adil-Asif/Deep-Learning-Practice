#Introduction to Numpy

```
#importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
#creating new array
a = np.array({1,2,3,4,5,6})
print(a)
print(type(a))
```

```
#initial value 0
#final value 60
#interval  2
a2= np.arange(1,60,2)
print(a2)
```

```
#create array with shape function
print(np.zeros(shape=[4,4]))
#same can be done for ones
print(np.ones(shape=[3,4]))
```

```
#0 and 100 are range 5 represents number of elements in array
rand= np.random.randint(0,100,5)
print(rand)
```

```
#5 random numbers represented in an array
randm= np.random.randn(5)
print(randm)
```

```
#max min in array 
print(rand.max())
print(rand.min())
```

```
#index of max value in array
print(rand.argmax())
print(rand.argmin())
```

```
#Reshape into matrix
mat = np.arange(0,20).reshape(4,5)
print(mat)
```

```
#slicing
print(mat[1:3])
print(mat[:,:1])
```



#Image Manipulation



```
#opening image
#cv2 already converts images to arrays directly when it is reading an image
img = cv2.imread("/content/Pic1.png")
cv2_imshow(img)
#type of img will be returned nonetype when you enter wrong image address in imread
print(type(img))
img.shape
```

```
#matplotlib --> RGB Channel
#cv2 --> BGR channel
#thus when plt tries to access img color changes are shown
plt.imshow(img)
```

```
#cvtColor= convert color
#syntax of cvtColor(image_name,color channel you want to change)
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```
grey_img = cv2.imread("/content/Pic1.png",0)
cv2_imshow(grey_img)
print(grey_img.shape)
```

```
#Image Resize
rsize = cv2.resize(fix_img,(1000,1000))
plt.imshow(rsize)
```

```
#Flip Image0, for flipping the image around the x-axis (vertical flipping);
# = 0, for flipping the image around the x-axis (vertical flipping);
# > 0 for flipping around the y-axis (horizontal flipping);
# < 0 for flipping around both axes.
flp = cv2.flip(fix_img,0)
plt.imshow(flp)
```

```
#creating a black image

black = np.zeros(shape=(450,450,3),dtype=np.int16)
plt.imshow(black)
```

```
#adding a rectangle to the image
cv2.rectangle(black, pt1=(150,150), pt2=(400,400) ,color=(0,250,0),thickness=7)
plt.imshow(black)
```

```
#adding circle to image
cv2.circle(black ,center=(150,150) ,radius=60 , color=(255,255,0) ,thickness = 6)
plt.imshow(black)
```

```
#merging circle and rectangle
cv2.line(black, pt1=(0,50), pt2= (450,50), color = (0,255,0), thickness= 7)
cv2.circle(black ,center=(150,150) ,radius=30 , color=(0,0,255) ,thickness = 6)
cv2.circle(black ,center=(150,150) ,radius=90 , color=(255,255,0) ,thickness = 6)
cv2.circle(black ,center=(150,150) ,radius=60 , color=(255,0,255) ,thickness = 6)
cv2.rectangle(black, pt1=(150,150), pt2=(400,400) ,color=(0,255,255),thickness=7)
plt.imshow(black)
```

```
font = cv2.FONT_HERSHEY_DUPLEX
cv2.putText(black, text="PATTERN",org=(155,350),fontScale= 1.8, fontFace=font, color=(255,255,255), thickness= 1)
plt.imshow(black)
```

```
blank = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blank)
```

```
#polygon
vertices = np.array([[[250,150],[100,300],[300,300],[200,100]]])
vertices.shape
cv2.polylines(blank,[vertices],color=(255,0,255),thickness=4,isClosed=True)
plt.imshow(blank) 
```
