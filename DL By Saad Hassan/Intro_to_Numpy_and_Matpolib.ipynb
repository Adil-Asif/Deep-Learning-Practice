> **Using NUMPY and it's functions for manipulation of Array**



```
#imshow contradicts with cv2
#import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
#creating new array
a = np.array({1,2,3,4,5,6})
print(a)
print(type(a))
```

```
#initial value 0
#final value 60
#interval  2
a2= np.arange(1,60,2)
print(a2)
```

```
#create array with shape function
print(np.zeros(shape=[4,4]))
#same can be done for ones
print(np.ones(shape=[3,4]))
```

```
#0 and 100 are range 5 represents number of elements in array
rand= np.random.randint(0,100,5)
print(rand)
```

```
#5 random numbers represented in an array
randm= np.random.randn(5)
print(randm)
```

```
#max min in array 
print(rand.max())
print(rand.min())
```

```
#index of max value in array
print(rand.argmax())
print(rand.argmin())
```

```
#Reshape into matrix
mat = np.arange(0,20).reshape(4,5)
print(mat)
```

```
#slicing
print(mat[1:3])
print(mat[:,:1])
```



> **Using Matpolib to import and image maniputlation**



```
from PIL import Image
#opens image and displays it
#image is always an array if it is not we need to covert it to an array
pic = Image.open("/content/Pic1.png")
```

```
pic
```

```
#show type of pic
print(type(pic))
```

```
#As pic is not an array we need to convert it
#converts the pic to type array
pics = np.asarray(pic)
print(type(pics))
```

```
#tells the shape and checks if image is rgb or only grey
print(pics.shape)
```

```
#Jet reads images more carefully it highlights details
#first element is height and width third is color channel
# 0 = R, 1 = G, 2 = B
#cmap applies color filter to image
plt.imshow(pics[:,:,0], cmap="jet")
```
